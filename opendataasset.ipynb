```python
file_path = r"C:\Users\daiyu\Documents\GitHub\2024colloquium_II\data\Local_Law_251_of_2017__Published_Data_Asset_Inventory_20241021.csv"
df = pd.read_csv(file_path, parse_dates=['Category'])
```

```python
category_counts = df['Category'].value_counts().sort_index()

# Display the counts
print(category_counts)

# Plot the bar graph
plt.figure(figsize=(20, 6))
category_counts.plot(kind='bar', color='lightgreen')

# Add titles and labels
plt.title('Number of dataset by category')
plt.xlabel('Category')
plt.ylabel('Number of Cases')

# Rotate x-axis labels for better readability
plt.xticks(rotation=30)

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Step 4: Inspect the Data
print(df.columns)  # Check the column names to identify the 'visits' column

# Step 5: Sort Data by Visits
# Assuming the column is called 'visits' or something similar
df_sorted = df.sort_values(by='Visits', ascending=False)

# Step 6: Extract the Top 10 Visited Datasets
top_assets = df_sorted[['Name', 'Visits']].head(10)  # Replace 'asset_name' with the correct column for dataset names

# Step 7: Plot the Bar Graph
plt.figure(figsize=(10, 6))  # Set the figure size
plt.barh(top_assets['Name'], top_assets['Visits'], color='blue')  # Horizontal bar graph
plt.xlabel('Number of Visits')
plt.ylabel('Dataset Name')
plt.title('Top 10 Most Visited NYC Open Data Datasets')

# Step 8: Show the Plot
plt.gca().invert_yaxis()  # Invert y-axis to have the most visited on top
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
